```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv(r"C:\Users\benan\Documents\Data_Scientist\Streamlit\wine_clustering_data.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.hist(figsize=(15, 12), bins=20, edgecolor='black')
plt.suptitle("Feature Distribution of Wine Dataset", fontsize=16)
plt.tight_layout()
plt.show()
```

```python

numerical_features = df.select_dtypes(include=['int64', 'float64'])

scaler = StandardScaler()

scaled_data = scaler.fit_transform(numerical_features)

df_scaled = pd.DataFrame(scaled_data, columns=numerical_features.columns)

df_scaled.head()
```

```python
dbscan = DBSCAN(eps=2, min_samples=2)
clusters = dbscan.fit_predict(df_scaled)

df['Cluster'] = clusters

df.head()
```

```python
df['Cluster'].value_counts()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(
    df['alcohol'],
    df['malic_acid'],
    c=df['Cluster'],
    cmap='viridis',
    s=60
)

plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.title("DBSCAN Clustering on Wine Dataset")
plt.colorbar(label="Cluster")
plt.show()
```

```python
eps_values = [1.5, 2, 3]
min_samples_values = [2, 4]

plt.figure(figsize=(12, 8))
plot_num = 1

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(df_scaled)

        plt.subplot(2, 3, plot_num)
        plt.scatter(
            df_scaled['alcohol'],
            df_scaled['malic_acid'],
            c=labels,
            cmap='viridis',
            s=40
        )
        plt.title(f"eps={eps}, min_samples={min_samples}")
        plot_num += 1

plt.tight_layout()
plt.show()
```
